```python
def unicodeReplace(text):
    replacements = {
        "òa": "oà", "óa": "oá", "ỏa": "oả", "õa": "oã", "ọa": "oạ",
        "òe": "oè", "óe": "oé", "ỏe": "oẻ", "õe": "oẽ", "ọe": "oẹ",
        "ùy": "uỳ", "úy": "uý", "ủy": "uỷ", "ũy": "uỹ", "ụy": "uỵ",
        "Ủy": "Uỷ", "\n": "." , "\t": "."  # Add more replacements as needed
    }
    for old, new in replacements.items():
        text = text.replace(old, new)
    return text
```

```python
# Tính Top-K Accuracy và F1-Score
def top_k_accuracy_f1(y_true, y_prob, k):
    top_k_predictions = np.argsort(y_prob, axis=1)[:, -k:]  # Lấy top-k chỉ số dự đoán
    correct_top_k = [y_true[i] in top_k_predictions[i] for i in range(len(y_true))]  # Kiểm tra ground truth trong top-k
    top_k_accuracy = np.mean(correct_top_k)  # Accuracy Top-K

    # Tính F1-Score Top-K (giả lập thành nhị phân, nếu ground truth có trong top-k là 1, ngược lại là 0)
    y_true_binary = np.ones(len(y_true))  # Ground truth luôn là "có trong top-k"
    y_pred_binary = np.array(correct_top_k).astype(int)  # 1 nếu có trong top-k, 0 nếu không
    top_k_f1 = f1_score(y_true_binary, y_pred_binary)

    return top_k_accuracy, top_k_f1
```

```python
def calculate_top5_accuracy(y_pred_top5, y_test):
    """
    Tính accuracy của y_pred_top5 với y_test dựa trên tiêu chí top 5 nhãn.

    Args:
        y_pred_top5 (list of dict): Danh sách các từ điển chứa nhãn và xác suất (top 5).
        y_test (list): Danh sách nhãn thực sự.

    Returns:
        float: Accuracy dựa trên top 5 nhãn.
    """
    correct = 0
    total = len(y_test)

    for i, true_label in enumerate(y_test):
        predicted_labels = list(y_pred_top5[i].keys())  # Lấy danh sách nhãn dự đoán (top 5)
        if true_label in predicted_labels:
            correct += 1

    return correct / total

# Tính accuracy
accuracy = calculate_top5_accuracy(y_pred_svm_top5, y_test)
print(f"Top-2 Accuracy: {accuracy * 100:.2f}%")
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np

# Tải mô hình
# model_name = "bert-base-uncased"
# model_name = "roberta-base"
# model_name = "roberta-large"
model_name = "distilbert-base-uncased"


tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Định nghĩa encoder cho BERT
class SentenceEncoder:
    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer

    def fit_transform(self, sentences):
        return self._encode(sentences)

    def transform(self, sentences):
        return self._encode(sentences)

    def _encode(self, sentences):
        embeddings = []
        for sentence in sentences:
            inputs = self.tokenizer(sentence, return_tensors="pt", truncation=True, padding="max_length", max_length=50)
            outputs = self.model(**inputs)
            sentence_embedding = outputs.last_hidden_state.mean(dim=1).detach().numpy()  # Lấy trung bình các token
            embeddings.append(sentence_embedding[0])
        return np.array(embeddings)

# Khởi tạo encoder
encoder = SentenceEncoder(model, tokenizer)

# Encode dữ liệu
X_train_encoder = encoder.fit_transform(X_train)
X_test_encoder = encoder.transform(X_test)
```

## **Phân tích tần suất các df**

```python
# Tạo mapping cho từng cột
def create_mapping(column_values):
    unique_values = column_values.unique()
    mapping = {value: f"{column_values.name[:3]}{i+1}" for i, value in enumerate(unique_values)}
    return mapping
```

### df_train

```python
df = df_train
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Mapping cho các cột
context_mapping = create_mapping(df['context'])
article_mapping = create_mapping(df['article'])
document_mapping = create_mapping(df['document'])

# Thay thế các giá trị trong DataFrame bằng mã mapping
df['context_mapped'] = df['context'].map(context_mapping)
df['article_mapped'] = df['article'].map(article_mapping)
df['document_mapped'] = df['document'].map(document_mapping)

# Thống kê tần suất
context_distribution = df['context'].value_counts().rename_axis('context').reset_index(name='count')
article_distribution = df['article'].value_counts().rename_axis('article').reset_index(name='count')
document_distribution = df['document'].value_counts().rename_axis('document').reset_index(name='count')

# Vẽ biểu đồ cho từng cột
def plot_distribution(distribution, mapping, column_name):
    plt.figure(figsize=(12, 6))
    # Thay thế giá trị dài bằng mã ngắn
    distribution['mapped'] = distribution[column_name].map(mapping)
    plt.bar(distribution['mapped'], distribution['count'], color='skyblue')
    plt.xlabel('Mapped Values')
    plt.ylabel('Frequency')
    plt.title(f"Distribution of {column_name}")
    # plt.xticks(rotation=90)
    plt.xticks([])
    plt.show()

# Vẽ biểu đồ cho 'context', 'article', 'document'
plot_distribution(context_distribution, context_mapping, 'context')
plot_distribution(article_distribution, article_mapping, 'article')
plot_distribution(document_distribution, document_mapping, 'document')

# Hiển thị bảng mapping
print("Mapping cho 'context':")
print(context_mapping)
print("\nMapping cho 'article':")
print(article_mapping)
print("\nMapping cho 'document':")
print(document_mapping)
```

## Độ dài

```python
import pandas as pd
import matplotlib.pyplot as plt

# Hàm tính độ dài chuỗi
def calculate_length_statistics(df, column_name):
    df[f"{column_name}_length"] = df[column_name].fillna('').apply(len)
    stats = df[f"{column_name}_length"].describe()  # Thống kê cơ bản
    return stats

# Hàm vẽ phân phối độ dài
def plot_length_distribution(df, column_name, dataset_name):
    plt.figure(figsize=(12, 6))
    plt.hist(df[f"{column_name}_length"], bins=30, color='skyblue', alpha=0.7, edgecolor='black')
    plt.title(f"Length Distribution of {column_name} in {dataset_name}")
    plt.xlabel("Length (number of characters)")
    plt.ylabel("Frequency")
    plt.show()

# Phân tích các DataFrame: df_train, df_test, df_val
datasets = {
    "Train": df_train,
    "Test": df_test,
    "Validation": df_val
}

# Cột cần phân tích
columns_to_analyze = ["question", "extractive answer", "abstractive answer"]

# Phân tích từng tập dữ liệu
for dataset_name, df in datasets.items():
    print(f"Dataset: {dataset_name}")
    for column in columns_to_analyze:
        # Tính thống kê
        stats = calculate_length_statistics(df, column)
        print(f"Statistics for {column} in {dataset_name}:")
        print(stats)

        # Vẽ phân phối độ dài
        plot_length_distribution(df, column, dataset_name)
```
